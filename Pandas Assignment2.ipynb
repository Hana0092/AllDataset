```python
# Imports
import pandas as pd
import requests
import numpy as np

# load the dataset
PATH_URL_3 = "https://archive.ics.uci.edu/ml/machine-learning-databases/flags/flag.data"

# Save column names as a list
col_list_3 = ["name", "landmass", "zone","area","population", "language", "religion", "bars", "stripes","colours","red","green","blue","gold","white",
              "black","orange","mainhue","circles","crosses","saltires","quarters","sunstars","crescent","triangle","icon","animate","text","topleft","botright"]

# Pass in col_ls into names parameter to explicitly name each column
df = pd.read_csv(PATH_URL_3, names=col_list_3)
```

- Use `loc` to slice your DataFrame for the 2 columns that returned the largest `sum`

```python
num_df = df.select_dtypes(include=np.number) # first get all numeric columns

names = num_df.sum().sort_values().tail(2)       # get the sum of the dataframe and then sort and take the last 2 values 

print("All the column names and their sums: \n",num_df.sum().sort_values())

print("\n"+"The 2 columns with the largest sum are: ")
df.loc[:,[names.index[0],names.index[1]]]      # get all rows for the names of the max sum values
```

- Use `iloc` to slice your DataFrame for the 2 columns that returned the largest `mean`

```python
num_df = df.select_dtypes(include=np.number) # first get all numeric columns

names = num_df.mean().sort_values().tail(2)       # get the mean of the dataframe and then sort and take the last 2 values 

print("All the column names and their means: \n",num_df.mean().sort_values())

print("\n"+"The 2 columns with the largest means are: ")
df.loc[:,[names.index[0],names.index[1]]]      # get all rows for the names of the largest mean values
```
